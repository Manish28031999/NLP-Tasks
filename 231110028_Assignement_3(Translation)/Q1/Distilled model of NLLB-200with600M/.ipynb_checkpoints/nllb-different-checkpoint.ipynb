```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

# article = "UN Chief says there is no military solution in Syria"
# inputs = tokenizer(article, return_tensors="pt")

# translated_tokens = model.generate(
#     **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["fra_Latn"], max_length=30
# )
# tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
```

```python
# here please specify the file path correctly

with open("/kaggle/input/devtest2021/test.en", "r", encoding="utf-8") as f:
    test_en = f.readlines()

with open("/kaggle/input/devtest2021/test.hi", "r", encoding="utf-8") as f:
    test_hi = f.readlines()

with open("/kaggle/input/devtest2021/test.gu", "r", encoding="utf-8") as f:
    test_gu = f.readlines()
```

```python
print(len(test_en))
print(len(test_hi))
print(len(test_gu))
```

```python
# select any 1000 random sentence

import numpy as np
indices = np.random.choice(len(test_en), 1000, replace=False)
```

```python
model = model.to('cuda')
```

# English to Hindi

```python
tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="eng_Latn"
)

translations_Hindi = []

# Translate the dev set
for i in indices:
  inputs = tokenizer(test_en[i].strip(), return_tensors="pt")
  inputs = {key: value.to(model.device) for key, value in inputs.items()}
  translated_tokens = model.generate(
      **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], max_length=100
  )
  hindi_translation = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  translations_Hindi.append(hindi_translation)




#   translations = []

# # Translate the dev set
# for i in indices:
#     inputs = tokenizer(dev_en[i].strip(), return_tensors="pt")
#     inputs = {key: value.to(model.device) for key, value in inputs.items()}
#     translated_tokens = model.generate(
#         **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], max_length=40
#     )
#     hindi_translation = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
#     translations.append(hindi_translation)
```

```python
# File path to save the text file
txt_file = 'NLLB_En_to_hi_Text_Final.txt'

# Open the text file in write mode
with open(txt_file, mode='w', encoding='utf-8') as file:
    # Write each sentence to the text file
    for sentence in translations_Hindi:
        file.write(sentence + '\n')
```

# Hindi to English

```python
translations_English = []

tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="hin_Deva"
)

# Translate the dev set
for i in indices:
  inputs = tokenizer(test_hi[i].strip(), return_tensors="pt")
  inputs = {key: value.to(model.device) for key, value in inputs.items()}
  translated_tokens = model.generate(
      **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["eng_Latn"], max_length=100
  )
  English_translation = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  translations_English.append(English_translation)
```

```python
# File path to save the text file
txt_file1 = 'NLLB_Hi_to_En_Text_Final.txt'

# Open the text file in write mode
with open(txt_file1, mode='w', encoding='utf-8') as file:
    # Write each sentence to the text file
    for sentence in translations_English:
        file.write(sentence + '\n')
```

# Hindi to Gujarati

```python
tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="hin_Deva"
)

translations_Gujarati = []

# Translate the dev set
for i in indices:
  inputs = tokenizer(test_hi[i].strip(), return_tensors="pt")
  inputs = {key: value.to(model.device) for key, value in inputs.items()}
  translated_tokens = model.generate(
      **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["guj_Gujr"], max_length=100
  )
  Gujarati_translation = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  translations_Gujarati.append(Gujarati_translation)
```

```python
# File path to save the text file
txt_file2 = 'NLLB_Hi_To_GU_Text_Final.txt'

# Open the text file in write mode
with open(txt_file2, mode='w', encoding='utf-8') as file:
    # Write each sentence to the text file
    for sentence in translations_Gujarati:
        file.write(sentence + '\n')
```

# Gujarati to Hindi

```python
tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="guj_Gujr"
)
translations_Hindi_from_Gujarati = []

# Translate the dev set
for i in indices:
  inputs = tokenizer(test_gu[i].strip(), return_tensors="pt")
  inputs = {key: value.to(model.device) for key, value in inputs.items()}
  translated_tokens = model.generate(
      **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], max_length=100
  )
  Hindi_translation_from_Gujrati = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  translations_Hindi_from_Gujarati.append(Hindi_translation_from_Gujrati)
```

```python
# File path to save the text file
txt_file3 = 'NLLB_GU_TO_HI_Text_Final.txt'

# Open the text file in write mode
with open(txt_file3, mode='w', encoding='utf-8') as file:
    # Write each sentence to the text file
    for sentence in translations_Hindi_from_Gujarati:
        file.write(sentence + '\n')
```
